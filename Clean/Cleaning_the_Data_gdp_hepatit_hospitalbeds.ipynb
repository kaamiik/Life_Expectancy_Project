## GDP_per_capita

```python
import pandas as pd
```

```python
df = pd.read_csv('GDP_per_capita.csv',index_col = 0)
df
```

```python
# only keeping countries
df1 = df.loc[:'Zimbabwe']
df1
```

```python
#changing type to float
df1 = df1.astype(float)
df1
```

```python
#filling missing values with interpolation
df2 = df1.interpolate(axis=1, limit_direction='both')
df2
```

```python
df2.to_csv('gdp_per_capita.csv')
```
