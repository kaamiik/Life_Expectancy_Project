# Literacy Rate

### Read and Clean data

```python
import pandas as pd

literacy_rate = pd.read_excel('literacy_rate.xlsx',index_col=0)
```

```python
literacy_rate 
```

```python
iso_code = literacy_rate["Alpha-3 code"]
iso_code
del literacy_rate["Alpha-3 code"] #Drop Iso Code
```

### Fill NaN values with mean

```python
import numpy as np
for columns in literacy_rate.columns:
    literacy_rate[columns].replace('..',np.nan, inplace=True)
```

```python
literacy_rate = literacy_rate.T
literacy_rate
```

```python
for col in literacy_rate:
    literacy_rate[col] = pd.to_numeric(literacy_rate[col], errors='coerce')
```

```python
import numpy as np
for col in literacy_rate:
    literacy_rate[col].replace(np.nan, literacy_rate[col].mean(), inplace=True)
```

```python
m = literacy_rate.mean(axis=1)
for i, col in enumerate(literacy_rate):
    # using i allows for duplicate columns
    # inplace *may* not always work here, so IMO the next line is preferred
    # df.iloc[:, i].fillna(m, inplace=True)
    literacy_rate.iloc[:, i] = literacy_rate.iloc[:, i].fillna(m)
```

```python
literacy_rate = literacy_rate.T
```

```python
literacy_rate
```

```python
literacy_rate.insert(loc=0, column="Iso_Code", value=iso_code) # Add Iso Code again
literacy_rate.insert(loc=0, column='Country', value=literacy_rate.index) # Add country Column
```

```python
literacy_rate = literacy_rate.melt(id_vars=["Country","Iso_Code"], 
        var_name="Year", 
        value_name="Literacy Rate")
literacy_rate
```

### Pivot the dataframe

```python
literacy_rate.drop_duplicates(inplace=True)
literacy_rate
```

# Merge Columns

```python
df0 = literacy_rate.merge(aids, how='inner', on=['Iso_Code','Year']).merge(overweight, how='inner', on=['Iso_Code','Year'])
#df1 = gdp.merge(hepatit, how='inner', on='Iso_Code').merge(hospital_beds, how='inner', on='Iso_Code')
#df0.merge(df1, how='inner', on='Iso_Code')
df0 = df0.drop(['Country_x','Country_y','Country'],axis=1)#'Year_x','Year_y'], axis=1)
#df0 = df0.rename(columns={'Country_x':'Country'})
df0
```

```python
df0.drop_duplicates(inplace=True)
df0
```

```python
df0['Year'] = df0['Year'].astype(int)
```

```python
df1 = gdp.merge(hepatit, how='inner', on=['Iso_Code','Year']).merge(hospital_beds, how='inner', on=['Iso_Code','Year'])
```

```python
df1 = df1.drop(['Country_y','Country_x','Country'], axis=1)
#df1 = df1.rename(columns={'Country_x':'Country'})
df1
```

```python
df1.drop_duplicates(inplace=True)
df1['Year'] = df1['Year'].astype(int)
df1
```

### Merge for estimate

```python
df = df0.merge(df1, how='inner', on=['Iso_Code','Year'])
df
```

### Merge for Prediction

```python
df_2022 = df[df['Year']==2022]
df_2022
```

```python
df = df[df["Year"] < 2022]
df
```
